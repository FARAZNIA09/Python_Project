```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("EPL_20_21.csv")
df
```

```python
gb = df.groupby("Nationality").agg({"Goals":'count'})
asc = gb.sort_values(by = "Goals",ascending =False)
ordered = asc.head(10)
pdd =pd.DataFrame(ordered)
pdd
```

```python
# plt.figure(figsize = [9,18])
# plt.yticks(rotation = 'vertical')
# float_number = float(pdd)
sb.heatmap(pdd, annot = True)
plt.show()
```

```python
gb2 = df.groupby("Name").agg({'Goals':'sum'})
maxfirst = gb2.sort_values(by="Goals", ascending = False)
plot = maxfirst.head(15)
plot
```

```python
plt.figure(figsize=[8,5])
plt.xticks(rotation = "vertical")
sb.set(font_scale = 1)
sb.lineplot(plot)
plt.show()
```

```python
Yellow_Card = df.groupby("Club").agg({"Yellow_Cards":"count"})
Yellow_Card
```

```python

plt.xticks(rotation = "vertical")
sb.scatterplot(Yellow_Card)
plt.show()
```
